```python
class FuzzyImageEnhancement:

    def __init__(self):
        self.intensity = None
        self.new_intensity = None
        self.modification = None
        pass

    def init_fuzzy_system(self, image, is_gray=False):
        # convert to LAB color system
        lab = image.copy()
        if not is_gray:
            lab = cv.cvtColor(image, cv.COLOR_RGB2LAB)

        # get the L channel
        l = lab[:,:,0]

        # caculate the mean intensity of the LAB image
        m = np.mean(l)
        if m < 128:
            m = 127 - (127 - m)/2
        else:
            m = 128 + m/2

        # initialize the input membership system
        self._init_input(m)

        # initialize the output membership system
        self._init_output(m)

        # initialize the rules and controller for the system to function
        self._init_rules(self.intensity, self.new_intensity)

        # Now we should apply the fuzzy transform on each
        # pixel of the image, but if the image was too big
        # it will take alot of time, so instead, we precompute
        # the transforms for each unique value and then map it
        # to the with the other pixels
        # Precompute the fuzzy transform
        x = list(range(-50,306))
        zipped = dict(zip(x,[self._transfer(i) for i in x]))

        # Apply the transform to l channel
        u, inv = np.unique(l, return_inverse = True)
        l = np.array([zipped[i] for i in u])[inv].reshape(l.shape)

        # Min-max scale the output L channel to fit (0, 255):
        min_inten = np.min(l)
        max_inten = np.max(l)
        if is_gray:
            temp = (l - min_inten)/(max_inten - min_inten) * 255
            lab[:, :, 0] = temp
            lab[:, :, 1] = temp
            lab[:, :, 2] = temp
        else:
            lab[:, :, 0] = (l - min_inten)/(max_inten - min_inten) * 255

        # return image to RGB color system
        if not is_gray:
            lab = cv.cvtColor(lab, cv.COLOR_LAB2RGB)
        return lab

    
    def _init_input(self, m):
        # deffine fuzzy system input
        intensity = ctrl.Antecedent(np.arange(-30, 306, 1), 'intensity')
        # initialize memberships
        intensity['extremelyDark'] = fuzz.gaussmf(intensity.universe, -50, m/6)
        intensity['veryDark'] = fuzz.gaussmf(intensity.universe, 0, m/6)
        intensity['dark'] = fuzz.gaussmf(intensity.universe, m/2, m/6)
        intensity['slightlyDark'] = fuzz.gaussmf(intensity.universe, 5*m/6, m/6)
        intensity['slightlyBright'] = fuzz.gaussmf(intensity.universe, m+(255-m)/6 ,(255-m)/6)
        intensity['bright'] = fuzz.gaussmf(intensity.universe, m+(255-m)/2, (255-m)/6)
        intensity['veryBright'] = fuzz.gaussmf(intensity.universe, 255, (255-m)/6)
        intensity['extremeBright'] = fuzz.gaussmf(intensity.universe, 305, (255-m)/6)
        # save as class variable
        self.intensity = intensity
    
    def _init_output(self, m):
        # deffine fuzzy system output
        new_intensity = ctrl.Consequent(np.arange(-50, 306, 1), 'new_intensity')
        # initialize memberships
        new_intensity['extremelyDark'] = fuzz.gaussmf(new_intensity.universe, -50, m/6)
        new_intensity['veryDark'] = fuzz.gaussmf(new_intensity.universe, 0, m/6)
        new_intensity['dark'] = fuzz.gaussmf(new_intensity.universe, m/2, m/6)
        new_intensity['slightlyDark'] = fuzz.gaussmf(new_intensity.universe, 5*m/6, m/6)
        new_intensity['slightlyBright'] = fuzz.gaussmf(new_intensity.universe, m+(255-m)/6 ,(255-m)/6)
        new_intensity['bright'] = fuzz.gaussmf(new_intensity.universe, m+(255-m)/2, (255-m)/6)
        new_intensity['veryBright'] = fuzz.gaussmf(new_intensity.universe, 255, (255-m)/6)
        new_intensity['extremelyBright'] = fuzz.gaussmf(new_intensity.universe, 305, (255-m)/6)
        # save as class variable
        self.new_intensity = new_intensity
    
    def _init_rules(self, intensity, new_intensity):
        rules = [None] * 6
        # deffine each rule in our fuzzy system
        rules[0] = ctrl.Rule(intensity['veryDark'], new_intensity['extremelyDark'])
        rules[1] = ctrl.Rule(intensity['dark'], new_intensity['veryDark'])
        rules[2] = ctrl.Rule(intensity['slightlyDark'], new_intensity['dark'])
        rules[3] = ctrl.Rule(intensity['slightlyBright'], new_intensity['bright'])
        rules[4] = ctrl.Rule(intensity['bright'], new_intensity['veryBright'])
        rules[5] = ctrl.Rule(intensity['veryBright'], new_intensity['extremelyBright'])
        # initialize the controller according to our rules 
        modification_ctrl = ctrl.ControlSystem(rules)
        modification = ctrl.ControlSystemSimulation(modification_ctrl)
         # save as class variable
        self.modification = modification
    
    def _transfer(self, x):
        self.modification.input['intensity'] = x
        self.modification.compute()
        return self.modification.output['new_intensity']
```

```python
def plot_image(image_path, is_gray):
    image = cv.imread(image_path)
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    
    system = FuzzyImageEnhancement()
    enhanced_image = system.init_fuzzy_system(image, is_gray=is_gray)
    
    fig = plt.figure(figsize=(10, 10))
    fig.add_subplot(2, 2, 1)
    plt.imshow(image)
    fig.add_subplot(2, 2, 2)
    plt.hist(image.ravel(), 256, [0, 256])
    fig.add_subplot(2, 2, 3)
    plt.imshow(enhanced_image)
    fig.add_subplot(2, 2, 4)
    plt.hist(enhanced_image.ravel(), 256, [0, 256])
    plt.show()
```
