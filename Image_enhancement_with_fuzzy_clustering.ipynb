```python
import numpy as np
import matplotlib.pyplot as plt
from Cmeans import c_means as cm
import timeit
import cv2 as cv
from collections import Counter
```

```python
class ImageHEWithFuzzyClusteringLocally:

    def __init__(self, adjust=True):
        self.number_of_clusters = None
        self.adjust = adjust
        pass

    # the main function for image enhancement
    def enhance(self, image, m=10, n=4, is_gray=False):
        # we transform the image into a vector
        ravel_image = image.reshape(image.shape[0] * image.shape[1], image.shape[2])
        if any(ravel_image[i][0] > 1 for i in range(ravel_image.shape[0])):
            ravel_image = ravel_image / 255.0
        
        # clustering image and dividing into chunks
        clusters, clustered_image, chunks = self.preprocess_image(image, ravel_image, m=m, n=n, 
                                                                  chunking=True, is_gray=is_gray)
        clustered_image = np.uint8(clustered_image * 255)
        
        # going through each chunk and perform histogram equalization
        equalized_chunks = {}
        cluster_num = 0
        for chunk in chunks:
            tmp = np.uint8(np.fromiter(chunk.values(), dtype=int))
            tmp = self.special_equalize(tmp)
            i = 0
            for key in chunk:
                chunk[key] = tmp[i]
                i += 1
            equalized_chunks.update(chunk)
            cluster_num += 1
        
        # if the image is colored, then we convert to HSV and apply changes on V value
        hsv_image = None
        if not is_gray:
            hsv_image = cv.cvtColor(image, cv.COLOR_RGB2HSV)

        # changing the pixels of main image to enhanced pixels
        equal_clustered_image = clustered_image.copy()
        for i in range(equal_clustered_image.shape[0]):
            for j in range(equal_clustered_image.shape[1]):
                equal_clustered_image[i, j] = equalized_chunks[(i, j)]
        
        # convert back to RGB
        if not is_gray:
            hsv_image[:, :, 2] = equal_clustered_image[:, :, 2]
            equal_clustered_image = cv.cvtColor(hsv_image, cv.COLOR_HSV2RGB)

        return clustered_image, equal_clustered_image

    # this function is responsible for clustering and chunking
    def preprocess_image(self, real_image, image_vector, m=10, n=4, chunking=False, is_gray=False):
        self.number_of_clusters = n
        # initialize clustering method
        algo = cm.FuzzyClustering(max_iter=m, n_clusters=n, method="Cmeans", debug=False)
        start = timeit.default_timer()
        # apply clustering
        clusters, membership = algo.fit(image_vector)
        stop = timeit.default_timer()
        print('Time: ', stop - start)

        index = membership.argmax(axis=1)
        
        # gather pixels in chunks according to each cluster collection
        processed_image = np.zeros((image_vector.shape[0], clusters.shape[1]))
        for i in range(index.shape[0]):
            processed_image[i] = clusters[index[i]]
        processed_image = processed_image.reshape(real_image.shape[0], real_image.shape[1], real_image.shape[2])

        hsv_image = real_image.copy()
        if not is_gray:
            hsv_image = cv.cvtColor(hsv_image, cv.COLOR_RGB2HSV)
        chunks = [None] * clusters.shape[0]
        for i in range(clusters.shape[0]):
            chunks[i] = {}
        if chunking:
            for i in range(processed_image.shape[0]):
                for j in range(processed_image.shape[1]):
                    chunks[index[i * processed_image.shape[1] + j]][(i, j)] = hsv_image[i, j, 2]
        return clusters, processed_image, chunks
    
    # this is the function that apply the main Histogram equalization formula
    def _transfer(self, pixels, value, start, end, m):
        n = np.sum(pixels[pixels < value])
        cdf = n / m
        return start + (end - start) * cdf

    # this function takes each chunk and apply the equalization on it
    def special_equalize(self, pixels):
        # calculate the range of the chunk
        start = min(pixels)
        end = max(pixels)
        if self.adjust:
            diff = end - start
            start -= int(diff / self.number_of_clusters)
            end += int(diff / self.number_of_clusters)
            start = 0 if start < 0 else start
            end = 255 if end > 255 else end
        
        # make duo of each value and the new value from equalization
        m = np.sum(pixels)
        x = list(range(start, end + 1))
        zipped = dict(zip(x, [self._transfer(pixels, i, start, end, m) for i in x]))

        u, inv = np.unique(pixels, return_inverse=True)
        pixels = np.array([zipped[i] for i in u])[inv].reshape(pixels.shape)
        return pixels
```

```python
# here we apply enhancement to image and plot it
def plot_image(image_path, m=20, n=4, is_gray=True, adjust=True):
    image = cv.imread(image_path, cv.IMREAD_COLOR)
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

    system = ImageHEWithFuzzyClusteringLocally(adjust=adjust)
    clustered_image, enhanced_image = system.enhance(image, m=m, n=n, is_gray=is_gray)


    fig = plt.figure(figsize=(10, 10))
    fig.add_subplot(3, 2, 1)
    plt.imshow(image)
    plt.title("real image")
    fig.add_subplot(3, 2, 2)
    plt.hist(image.ravel(), 256, [0, 256])
    fig.add_subplot(3, 2, 3)
    plt.imshow(clustered_image)
    plt.title("clustered image")
    fig.add_subplot(3, 2, 4)
    plt.hist(clustered_image.ravel(), 256, [0, 256])
    fig.add_subplot(3, 2, 5)
    plt.imshow(enhanced_image)
    plt.title("enhanced image")
    fig.add_subplot(3, 2, 6)
    plt.hist(enhanced_image.ravel(), 256, [0, 256])

    plt.show()
```

```python
plot_image('gentelman.jpg', is_gray = True, n=3, adjust=False)
```
