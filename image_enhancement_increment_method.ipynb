```python
class FuzzyImageEnhancement:

    def __init__(self):
        self.intensity = None
        self.adjustment = None
        self.modification = None
        pass

    def init_fuzzy_system(self, image, is_gray=False):
        # convert to HSV color system
        hsv = image.copy()
        if not is_gray:
            hsv = cv.cvtColor(image, cv.COLOR_RGB2HSV)

        # get the V channel
        v = hsv[:,:,2]

        # initialize the input membership system
        self._init_input()

        # initialize the output membership system
        self._init_output()

        # initialize the rules and controller for the system to function
        self._init_rules(self.intensity, self.adjustment)

        # Now we should apply the fuzzy transform on each
        # pixel of the image, but if the image was too big
        # it will take alot of time, so instead, we precompute
        # the transforms for each unique value and then map it
        # to the with the other pixels
        # Precompute the fuzzy transform
        x = list(range(0,256))
        zipped = dict(zip(x,[self._transfer(i) for i in x]))

        # Apply the transform to V channel
        u, inv = np.unique(v, return_inverse = True)
        v = np.array([zipped[i] for i in u])[inv].reshape(v.shape)

        # apply modification to actual image:
        if is_gray:
            hsv[:, :, 0] = v
            hsv[:, :, 1] = v
            hsv[:, :, 2] = v
        else:
            hsv[:, :, 2] = v

        # return image to RGB color system
        if not is_gray:
            hsv = cv.cvtColor(hsv, cv.COLOR_HSV2RGB)
        return hsv

    
    def _init_input(self):
        # deffine fuzzy system input
        intensity = ctrl.Antecedent(np.arange(0, 256, 1), 'intensity')
        # initialize memberships
        intensity['veryDark'] = fuzz.trapmf(intensity.universe, [0, 0,20 , 45])
        intensity['dark'] = fuzz.gaussmf(intensity.universe, 50, 10)
        intensity['darkGray'] = fuzz.gaussmf(intensity.universe, 80, 12)
        intensity['gray'] = fuzz.gaussmf(intensity.universe, 115 ,8)
        intensity['lightGray'] = fuzz.gaussmf(intensity.universe, 155 ,14)
        intensity['bright'] = fuzz.gaussmf(intensity.universe, 186, 12)
        intensity['veryBright'] = fuzz.trapmf(intensity.universe, [200,230,255,255])
        # save as class variable
        self.intensity = intensity
    
    def _init_output(self):
        # deffine fuzzy system output
        adjustment = ctrl.Consequent(np.arange(-50, 50, 1), 'adjustment')
        # initialize memberships
        adjustment['darker'] = fuzz.gaussmf(adjustment.universe, -50, 8)
        adjustment['slightlyDarker'] = fuzz.gaussmf(adjustment.universe, -30, 8)
        adjustment['noChange'] = fuzz.gaussmf(adjustment.universe, 0, 4)
        adjustment['slightlyBright'] = fuzz.gaussmf(adjustment.universe, 12 ,6)
        adjustment['brighter'] = fuzz.gaussmf(adjustment.universe, 50 ,8)
        # save as class variable
        self.adjustment = adjustment
    
    def _init_rules(self, intensity, adjustment):
        rules = [None] * 7
        # deffine each rule in our fuzzy system
        rules[0] = ctrl.Rule(intensity['veryDark'], adjustment['slightlyDarker'])
        rules[1] = ctrl.Rule(intensity['darkGray'], adjustment['slightlyDarker'])
        rules[2] = ctrl.Rule(intensity['gray'], adjustment['slightlyDarker'])
        rules[3] = ctrl.Rule(intensity['bright'], adjustment['slightlyBright'])
        rules[4] = ctrl.Rule(intensity['dark'], adjustment['darker'])
        rules[5] = ctrl.Rule(intensity['veryBright'], adjustment['noChange'])
        rules[6] = ctrl.Rule(intensity['lightGray'], adjustment['slightlyDarker'])
        # initialize the controller according to our rules 
        modification_ctrl = ctrl.ControlSystem(rules)
        modification = ctrl.ControlSystemSimulation(modification_ctrl)
         # save as class variable
        self.modification = modification
    
    def _transfer(self, x):
        self.modification.input['intensity'] = x
        self.modification.compute()
        temp = int(self.modification.output['adjustment']) + int(x)
        temp = 255 if temp > 255 else temp
        temp = 0 if temp < 0 else temp
        return temp
```

```python
def plot_image(image_path, is_gray):
    image = cv.imread(image_path)
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    
    system = FuzzyImageEnhancement()
    enhanced_image = system.init_fuzzy_system(image, is_gray=is_gray)
    
    fig = plt.figure(figsize=(10, 10))
    fig.add_subplot(2, 2, 1)
    plt.imshow(image)
    fig.add_subplot(2, 2, 2)
    plt.hist(image.ravel(), 256, [0, 256])
    fig.add_subplot(2, 2, 3)
    plt.imshow(enhanced_image)
    fig.add_subplot(2, 2, 4)
    plt.hist(enhanced_image.ravel(), 256, [0, 256])
    plt.show()
```

```python
plot_image('gentelman.jpg', is_gray = True)
```
